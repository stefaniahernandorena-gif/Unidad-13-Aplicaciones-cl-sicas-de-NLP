```python
# ==============================================================================
# UNIDAD 13: APLICACIONES CLÁSICAS DE NLP (VERSIÓN CORREGIDA Y OPTIMIZADA)
# ==============================================================================

# Instalación de librerías base actualizadas
!pip install -q transformers torch pandas matplotlib seaborn

import pandas as pd
import torch
from transformers import pipeline

# Configuración de hardware para los Transformers
device = 0 if torch.cuda.is_available() else -1

# ------------------------------------------------------------------------------
# 🔹 ESCENARIO 1: ANÁLISIS DE SENTIMIENTO (SOLUCIÓN MULTILINGÜE REAL)
# ------------------------------------------------------------------------------
print("Executing: Scenario 1 (Sentiment Analysis)...")

# REGLA DE INGENIERÍA: Reemplazamos TextBlob (obsoleto para español) por el Transformer multilingüe
sentiment_pipeline = pipeline(
    task="sentiment-analysis",
    model="cardiffnlp/twitter-xlm-roberta-base-sentiment",
    device=device
)

opiniones = [
    "El nuevo plan es excelente y muy económico",
    "La señal es mala y el servicio es lento",
    "No está mal, pero esperaba algo mejor",
    "Muy conforme con la atención al cliente"
]

print("\n📊 RESULTADOS ESCENARIO 1:")
for texto in opiniones:
    pred = sentiment_pipeline(texto)[0]
    # Mapeo visual para facilitar la interpretación del negocio
    label = pred["label"].upper()
    label_es = "POSITIVO" if "POS" in label else "NEGATIVO" if "NEG" in label else "NEUTRAL"
    print(f"Texto: '{texto}' ➔ Sentimiento: {label_es} (Confianza: {pred['score']:.2f})")


# ------------------------------------------------------------------------------
# 🔹 ESCENARIO 2: CLASIFICACIÓN DE TEXTO (REGLAS Y NUEVOS CASOS)
# ------------------------------------------------------------------------------
print("\n" + "="*70 + "\nExecuting: Scenario 2 (Text Classification)...")

def clasificar_noticia(texto):
    """Clasificador heurístico basado en reglas de negocio."""
    texto_lower = texto.lower()
    if any(p in texto_lower for p in ["dólar", "inflación", "bonos", "mercado", "suben", "consumo"]):
        return "economía"
    elif any(p in texto_lower for p in ["equipo", "partido", "campeonato", "delantero", "ganó"]):
        return "deportes"
    else:
        return "otros"

# REQUISITO DE ACTIVIDAD: Probando el clasificador con nuevos textos
nuevos_articulos = [
    "El partido fue intenso y muy disputado", # Ejemplo base
    "El banco central intervino el mercado para frenar la inflación", # Nuevo Economía
    "El delantero entrenó diferenciado antes de la final del campeonato" # Nuevo Deportes
]

print("\n📰 RESULTADOS ESCENARIO 2:")
for art in nuevos_articulos:
    print(f"Artículo: '{art}' ➔ Categoría Asignada: {clasificar_noticia(art).upper()}")


# ------------------------------------------------------------------------------
# 🔹 ESCENARIO 3: RESUMEN AUTOMÁTICO (SOLUCIÓN FRECUENCIAL SIN GENSIM ROTO)
# ------------------------------------------------------------------------------
print("\n" + "="*70 + "\nExecuting: Scenario 3 (Text Summarization)...")

def resumen_extractivo_frecuencial(texto, num_oraciones=2):
    """
    Alternativa robusta de ingeniería a Gensim.
    Calcula un resumen extractivo basado en la frecuencia de las palabras.
    """
    oraciones = [o.strip() for o in texto.split('\n') if o.strip()]
    palabras = texto.lower().split()
    frecuencia = Counter([p for p in palabras if len(p) > 4]) # Filtramos conectores cortos

    # Puntuamos cada oración según la importancia de sus palabras
    puntuacion_oraciones = {}
    for i, oracion in enumerate(oraciones):
        puntuacion_oraciones[i] = sum(frecuencia[p] for p in oracion.lower().split())

    # Seleccionamos las oraciones con mayor puntaje manteniendo el orden original
    indices_top = sorted(puntuacion_oraciones, key=puntuacion_oraciones.get, reverse=True)[:num_oraciones]
    resumen = " ".join([oraciones[idx] for idx in sorted(indices_top)])
    return resumen

from collections import Counter

texto_largo = """
El procesamiento de lenguaje natural es una rama de la inteligencia artificial que se enfoca en la interacción entre las computadoras y el lenguaje humano.
Permite analizar grandes volúmenes de texto, extraer información relevante y automatizar tareas como clasificación, resumen y análisis de sentimiento.
Esto reduce drásticamente los tiempos operativos y permite a los analistas enfocarse en la toma de decisiones estratégicas.
"""

print("\n📝 RESULTADOS ESCENARIO 3:")
print(f"Texto Original ({len(texto_largo.split())} palabras):\n{texto_largo.strip()}")
resumen_final = resumen_extractivo_frecuencial(texto_largo, num_oraciones=1)
print(f"\nResumen Automático Generado ({len(resumen_final.split())} palabras):\n{resumen_final}")


# ------------------------------------------------------------------------------
# 🔹 ESCENARIO 4: AUTOMATIZACIÓN DE SOPORTE (CHATBOTS)
# ------------------------------------------------------------------------------
print("\n" + "="*70 + "\nExecuting: Scenario 4 (Automated Chatbot)...")

faq = {
    "horario": "Nuestro horario de atención es de 9 a 18 hs.",
    "precio": "Podés consultar los precios en nuestra web.",
    "soporte": "Podés contactarte con soporte desde la app."
}

def chatbot(pregunta):
    pregunta_lower = pregunta.lower()
    for clave, respuesta in faq.items():
        if clave in pregunta_lower:
            return respuesta
    return "No pude entender tu consulta. ¿Podrías reformularla usando palabras clave como 'horario', 'precio' o 'soporte'?"

preguntas_usuario = [
    "¿Cuál es el horario de las oficinas?",
    "Quiero saber el precio del servicio",
    "Necesito soporte urgente",
    "Hola, ¿hay alguien ahí?"
]

print("\n🤖 RESULTADOS ESCENARIO 4:")
for preg in preguntas_usuario:
    print(f"Usuario: '{preg}' ➔ Bot: {chatbot(preg)}")
```

### ✍️ Resolución de Actividad - Escenario 1

#### 1. Identificación y Clasificación de Sentimientos:
* **"El nuevo plan es excelente y muy económico"** ➔ **POSITIVO** (Polaridad y carga emocional alta hacia el valor/precio).
* **"La señal es mala y el servicio es lento"** ➔ **NEGATIVO** (Detalla fallas críticas en la infraestructura técnica).
* **"No está mal, pero esperaba algo mejor"** ➔ **NEUTRAL / MIXTO** (Estructura adversativa que equilibra la balanza de satisfacción).
* **"Muy conforme con la atención al cliente"** ➔ **POSITIVO** (Validación directa sobre el factor humano del servicio).

Para el equipo de **Marketing**, permite medir el pulso de la marca en tiempo real, identificando detractores para contener crisis de reputación antes de que escalen en redes sociales.

### ✍️ Resolución de Actividad - Escenario 2

#### 1. Resultados de Nuevos Textos:
El clasificador fue expandido y validado con estructuras complejas como *"El banco central intervino el mercado para frenar la inflación"*, asignando correctamente la categoría **ECONOMÍA** gracias al rastreo de palabras clave.

#### 2. Utilidad de Técnicas de Reglas antes de Modelos Avanzados:
Implementar un clasificador heurístico (basado en reglas duras de `if/else`) es una práctica fundamental de ingeniería por tres razones:
1. **Solución al problema de arranque en frío (Cold Start):** Permite tener un sistema funcional cuando no se dispone de un dataset histórico etiquetado para entrenar una IA.
2. **Línea de Base (Baseline):** Funciona como la métrica mínima contra la cual comparar modelos complejos (si un Transformer pesado no supera la precisión de las reglas simples, no se justifica su costo de infraestructura).
3. **Interpretabilidad Absoluta:** No es una caja negra; el negocio sabe exactamente por qué una noticia fue asignada a una categoría.
